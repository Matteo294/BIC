## Task 1

### a) 
The human brain consumes a power $P$ of 20W, is composed of $N_n = 10^{11}$ neurons that fires with a frequency $\nu$ of 1Hz, and $N_s = 10^{15}$ synapses. Thus, the energy consuption per action potential is:
\begin{equation}
E_\text{act} = \frac{P}{N_n \nu} = 2^{-10} \text{J},
\end{equation}
and the energy consuption per synaptic event is:
\begin{equation}
E_\text{syn} = \frac{P}{N_\text{s} \nu} = 2^{-14} \text{J}.
\end{equation}

### b)
The energy consuption of the simulation per action potential is:
\begin{equation}
E_\text{act,K} = \frac{Pt}{N_n \nu} = 2.6 \text{J},
\end{equation}
and the energy consuption per synaptic event is:
\begin{equation}
E_\text{syn,K} = \frac{P}{N_n N_\text{connection} \nu} = 0.004\text{J}.
\end{equation}

### c)
If we scale the K simulation to the brain sixe in term of spikes we obtain a consuption of
\begin{equation}
P_\text{K,spikes} = P_K\frac{N_\text{neur,brain}}{N_\text{neur,K} } = 1.02\text{GW}.
\end{equation}
If we now scale it per number of synapses we get
\begin{equation}
P_\text{K,syn} = P_K\frac{N_\text{syn,brain}}{N_\text{neur,k} N\text{connections,k}} = 1.7\text{GW}.
\end{equation}

### d)
If we could speed  up the K supercomputer maintaining the energy/flop constant, to simulate a human brain (in term of spikes) we would need:
\begin{equation}
P_\text{K,speeded} = P_\text{K,spikes}\frac{t_\text{k}}{t_\text{brain}} = 2.4\times 10^{12}\text{W}.
\end{equation}
To do so, I would need approximately the power generated by 1375 nuclear power plants.
